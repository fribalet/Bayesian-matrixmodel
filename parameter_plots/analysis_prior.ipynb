```r
models <- c('m1p','m2p','m3p','m4p','m5p','m6p','m7p','m8p','m9p','m10p')

D <- list()
for(i in 1:length(models)){
    D[[i]] <- read.csv(paste('prior_',models[i],'.csv',sep=''))
}
```

```r
#pars <- c('gamma_max','respiration','E_star','w_ini_mu','w_ini_sigma','divrate')
pars  <- c('divrate','gamma_max')
nsamp <- dim(D[[1]])[1]

PARS <- array(NA,dim=c(nsamp,length(pars),length(models)))
for(j in 1:length(models)){
    parsmat           <- matrix(NA,nsamp,length(pars))
    colnames(parsmat) <- pars
    for(i in 1:length(pars)){
        parsmat[,i]   <- D[[j]][,grep(pars[i],colnames(D[[j]]))]
    }
    PARS[,,j] <- parsmat
}
```

```r
options(repr.plot.width=8, repr.plot.height=8)

i <- which(pars=='divrate')
par(mfrow=c(4,3),mar=c(2,2,2,2),oma=c(2,2,2,2))
for(j in 1:length(models)){
        hist(PARS[,i,j],main='',,breaks=20,xlab='',freq=TRUE,col=adjustcolor(j,alpha.f=0.5),ylim=c(0,1000))#,add=TRUE)
        mtext(models[j])
    }
mtext('Division Rate',outer=TRUE)
```

```r
options(repr.plot.width=8, repr.plot.height=8)

i <- which(pars=='gamma_max')
par(mfrow=c(4,3),mar=c(2,2,2,2),oma=c(2,2,2,2))
for(j in 1:length(models)){
        hist(PARS[,i,j],main='',,breaks=20,xlab='',freq=TRUE,col=adjustcolor(j,alpha.f=0.5),ylim=c(0,1000))#,add=TRUE)
        mtext(models[j])
    }
mtext(pars[i],outer=TRUE)
```

```r
options(repr.plot.width=8, repr.plot.height=8)

iplot <- sample(1:nsamp,size=500)

par(mfrow=c(4,3),mar=c(2,2,2,2),oma=c(2,2,2,2))
for(j in 1:length(models)){
        plot(PARS[iplot,2,j],PARS[iplot,1,j],xlab='',xlim=c(0,10),ylim=c(0,5),pch=1,cex=0.4,lwd=0.5,bty='n')
        mtext(models[j])
}
mtext('Gamma Max',side=1,outer=TRUE)
mtext('Division Rate',side=2,outer=TRUE)
```

```r
models2 <- c('m4p','m5p','m6p','m7p','m8p','m9p','m10p')
pars2   <- c('divrate','gamma_max','rho_max')
nsamp   <- dim(D[[1]])[1]

PARS2 <- array(NA,dim=c(nsamp,length(pars2),length(models2)))
for(j in 1:length(models2)){
    parsmat           <- matrix(NA,nsamp,length(pars2))
    colnames(parsmat) <- pars2
    for(i in 1:length(pars2)){
        parsmat[,i]   <- D[[j+3]][,grep(pars2[i],colnames(D[[j+3]]))]
    }
    PARS2[,,j] <- parsmat
}
```

```r
options(repr.plot.width=8, repr.plot.height=6)
iplot <- sample(1:nsamp,size=500)

par(mfrow=c(3,3),mar=c(2,2,2,2),oma=c(2,2,2,2))
for(j in 1:length(models2)){
        plot(PARS2[iplot,2,j],PARS2[iplot,3,j],xlab='',xlim=c(0,10),ylim=c(0,4),pch=1,cex=0.4,lwd=0.5,bty='n')
        mtext(models2[j])
}
mtext('Gamma Max',side=1,outer=TRUE)
mtext('Rho max',side=2,outer=TRUE)
```
