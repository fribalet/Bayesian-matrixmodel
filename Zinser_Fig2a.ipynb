```python
import pandas

data_fig2a = pandas.read_csv('data/Zinser_Figure2A.csv')

t = data_fig2a['exper time'] * 60
a_norm = data_fig2a['cells A'].copy()
a_norm /= a_norm[0]
b_norm = data_fig2a['cells B'].copy()
b_norm /= b_norm[0]

data_fig2a
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12,10))
ax.plot(t, a_norm, color='black', marker='s', label='normalized "cells A"')
ax.plot(t, b_norm, color='red', marker='^', label='normalized "cells B"')
for iday in range(2):
    ax.axvspan(iday*24*60+12*60, iday*24*60+22*60, color='0.7', zorder=0)
ax.legend()
None
```
