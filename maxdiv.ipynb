## a smoother light-dependent division rate

```python
import numpy as np

E = np.linspace(0.0, 2200.0, 200)

def maxdiv(E, thresh, sigma):
    d = np.exp(-((E-thresh)/sigma)**2)
    d[E<thresh] = 1.0
    return d
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,9))
for isample in range(100):
    thresh = np.random.uniform(low=0.0, high=2000.0)
    sigma = np.random.exponential(100)
    ax.plot(E, maxdiv(E, thresh, sigma), color='black', alpha=0.2)

```
